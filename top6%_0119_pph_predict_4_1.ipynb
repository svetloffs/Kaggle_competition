### func plot_learning_curve

```
# learning_curve(estimator, X, y, groups=None, train_sizes=ndarray with shape (5,), 
#                cv=None, scoring=None, exploit_incremental_learning=False, 
#                n_jobs=None, pre_dispatch='all', verbose=0, shuffle=False, 
#                random_state=None, error_score=nan, return_times=False,,)
def plot_learning_curve(estimator, title, X, y, ylim = None, 
                        cv=None, n_jobs = 1, train_sizes = np.linspace(.1, 1.0, 5)):
    plt.figure() 
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples") 
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(estimator,
                                                            X, y, n_jobs=n_jobs,
                                                            train_sizes=train_sizes, 
                                                            )
    train_scores_mean = np.mean(train_scores, axis=1) 
    train_scores_std = np.std(train_scores, axis=1) 
    test_scores_mean = np.mean(test_scores, axis=1) 
    test_scores_std = np.std(test_scores, axis=1) 
    plt.grid() 
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color="r") 
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std,
                     alpha=0.1, color="g") 
    plt.plot(train_sizes, train_scores_mean, 'r^-', 
             color="r", label="Training score") 
    plt.plot(train_sizes, test_scores_mean, 'o-', 
             color="g", label="Cross-validation score") 
    plt.legend(loc="best") 
    return plt
```

### func histogramm_build

```
def histogramm_build(df):
    try:
        df.name
        title = df.name
    except:
        title = df.columns
    sns.distplot(df, fit=norm, );
    # Get the fitted parameters used by the function
    (mu, sigma) = norm.fit(df);
    print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))
    #Now plot the distribution
    plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
                loc='best')
    plt.ylabel('Frequency')
    plt.title(title)
    #Get also the QQ-plot
    fig = plt.figure();
    res = stats.probplot(df, plot=plt);
    plt.show();
```

### func skew

```
def skew_data(df, num_features):
    # Check the skew of all numerical features
    skewed_feats = df[num_features].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
    print("\nSkew in numerical features: \n")
    skewness = pd.DataFrame({'Skew' :skewed_feats})
    skewness = skewness[abs(skewness) > 0.75] #0.75 - old
    print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))
    skewed_features = skewness.index
    lam = 0.15 #0.15
    for feat in skewed_features:
        #all_data[feat] += 1
        df[feat] = boxcox1p(df[feat], lam)
    return df
```

```
n_folds = 5 #5
history_score = list()
def rmsle_cv(model, title):
    cv = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(X.values)
    # cv = StratifiedKFold(n_folds, shuffle=True, random_state=42).get_n_splits(X.values)
    score = (-cross_val_score(model, X.values, y, scoring="neg_mean_squared_error", cv = cv))
    scores = cross_val_score(model, X.values, y, cv = cv)
    y_pred = cross_val_predict(model, X.values, y, cv=cv)
    mae = mean_absolute_error(np.expm1(y_pred), np.expm1(y))
    # r2 = r2_score(y, y_rmsle_pred)
    # name_model = str(model)[:str(model).find("(al")].split()[-1]
    # name_model = [k for k, v in locals().items() if v is model][0].capitalize()
    name_model = title
    print("{}:".format(name_model))
    print("Accuracy: {:.3%} (+/- {:.3%})".format(scores.mean(), scores.std() * 2))
    print("MSE-score: {:.4f} (+/-{:.4f})".format((np.sqrt(score)).mean(), (np.sqrt(score)).std()))
    print("MAE-score: {:.2f}".format(mae))
    print("Yt_pred: ", np.around(np.expm1(y_pred[:3]),1))
    # print(np.around(np.expm1(y_pred[:3]),1))
    # print("R2-score: {:.3f}".format(r2))
    MSE = str('MSE:') + str(round((np.sqrt(score)).mean(), 4))
    MAE = str('MAE:') + str(round(mae, 1))
    history_score.append((str(name_model), MSE, MAE))
    return history_score
```

```
class AveragingModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, models):
        self.models = models
        
    # we define clones of the original models to fit the data in
    def fit(self, X, y):
        self.models_ = [clone(x) for x in self.models]
        
        # Train cloned base models
        for model in self.models_:
            model.fit(X, y)

        return self
    
    #Now we do the predictions for cloned models and average them
    def predict(self, X):
        predictions = np.column_stack([
            model.predict(X) for model in self.models_
        ])
        return np.mean(predictions, axis=1)
```

```
class StackingAveragedModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, base_models, meta_model, n_folds=5):
        self.base_models = base_models
        self.meta_model = meta_model
        self.n_folds = n_folds
   
    # We again fit the data on clones of the original models
    def fit(self, X, y):
        self.base_models_ = [list() for x in self.base_models]
        self.meta_model_ = clone(self.meta_model)
        kfold = KFold(n_splits=self.n_folds, shuffle=True, random_state=156)
        
        # Train cloned base models then create out-of-fold predictions
        # that are needed to train the cloned meta-model
        out_of_fold_predictions = np.zeros((X.shape[0], len(self.base_models)))
        for i, model in enumerate(self.base_models):
            for train_index, holdout_index in kfold.split(X, y):
                instance = clone(model)
                self.base_models_[i].append(instance)
                instance.fit(X[train_index], y[train_index])
                y_pred = instance.predict(X[holdout_index])
                out_of_fold_predictions[holdout_index, i] = y_pred
                
        # Now train the cloned  meta-model using the out-of-fold predictions as new feature
        self.meta_model_.fit(out_of_fold_predictions, y)
        return self
   
    #Do the predictions of all base models on the test data and use the averaged predictions as 
    #meta-features for the final prediction which is done by the meta-model
    def predict(self, X):
        meta_features = np.column_stack([
            np.column_stack([model.predict(X) for model in base_models]).mean(axis=1)
            for base_models in self.base_models_ ])
        return self.meta_model_.predict(meta_features)
```
